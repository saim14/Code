```python
import matplotlib.pyplot as plt
import numpy as np

# Define the quadratic function
def quadratic_function(t):
    return 30 + 4*t + 2*t**2

# Generate values for t
t_values = np.linspace(-10, 10, 400)
y_values = quadratic_function(t_values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(t_values, y_values, label='$y = 30 + 4t + 2t^2$', color='blue')
plt.title('Graph of $y = 30 + 4t + 2t^2$')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```
