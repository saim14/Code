```python
import matplotlib.pyplot as plt
import numpy as np

# Define the quadratic function
def quadratic_function(t):
    return 30 + 4*t + 2*t**2

# Generate values for t
t_values = np.linspace(-10, 10, 400)
y_values = quadratic_function(t_values)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(t_values, y_values, label='$y = 30 + 4t + 2t^2$', color='blue')
plt.title('Graph of $y = 30 + 4t + 2t^2$')
plt.xlabel('t')
plt.ylabel('y')
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parametric equations for the heart shape
def heart_shape(t):
    x = 16 * np.sin(t)**3
    y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
    return x, y

# Generate values for t
t_values = np.linspace(0, 2 * np.pi, 1000)
x_values, y_values = heart_shape(t_values)

# Create the plot
plt.figure(figsize=(8, 8))
plt.plot(x_values, y_values, color='red')
plt.title('Heart Shape')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # Ensure the aspect ratio is equal
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parametric equations for the heart shape
def heart_shape(t):
    x = 16 * np.sin(t)**3
    y = 13 * np.cos(t) - 5 * np.cos(2 * t) - 2 * np.cos(3 * t) - np.cos(4 * t)
    return x, y

# Generate values for t
t_values = np.linspace(0, 2 * np.pi, 1000)
x_values, y_values = heart_shape(t_values)

# Create the plot
plt.figure(figsize=(8, 8))
plt.fill(x_values, y_values, color='red')  # Fill the heart shape with red color
plt.plot(x_values, y_values, color='black')  # Outline the heart shape
plt.title('Heart Shape')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')  # Ensure the aspect ratio is equal
plt.show()
```
